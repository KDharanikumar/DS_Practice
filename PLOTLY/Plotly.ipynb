```python
import plotly.express as px
import pandas as pd
# Load the Gapminder dataset
df = px.data.gapminder()
# Group by country and calculate the mean GDP per capita, then reset the index
grouped_df = df.groupby("country")['gdpPercap'].mean().reset_index()
# Optionally, sort the DataFrame by average GDP per capita
grouped_df = grouped_df.sort_values('gdpPercap', ascending=False)
# Create the bar chart
fig = px.bar(grouped_df, x='country', y='gdpPercap',
             labels={'gdpPercap': 'Average GDP per Capita'},
             title='Average GDP per Capita by Country')

# Show the figure
fig.show()
```

```python
from dash import Dash
from dash import dcc
from dash import html
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html, Input, Output

df = px.data.tips()
days = df.day.unique()
app = Dash()
#app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id="dropdown",
        options=[{"label": x, "value": x} for x in days],
    ),
    dcc.Graph(id="bar-chart"),
])

@app.callback(
    Output("bar-chart", "figure"),
    [Input("dropdown", "value")])

def update_bar_chart(day12):
    filtered = df["day"] == day12
    fig = px.bar(df[filtered], x="sex", y="total_bill",
                 color="smoker", barmode="group")
    return fig

app.run_server(debug=True)
```

```python
from dash import Dash
from dash import dcc
from dash import html
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html, Input, Output

df = px.data.iris()

#app = dash.Dash(__name__)
app = Dash(__name__)

app.layout = html.Div([
   dcc.Graph(id="scatter-plot"),
   html.P("Petal Width:"),
   dcc.RangeSlider(
       id='range-slider',
       min=0, max=2.5, step=0.1,
       marks={0: '0', 2.5: '2.5'},
       value=[0.5, 2]
   ),
])

@app.callback(
   Output("scatter-plot", "figure"),
   [Input("range-slider", "value")])
def update_bar_chart(slider_range):
   print(slider_range)
   low, high = slider_range
   mask = (df['petal_width'] > low) & (df['petal_width'] < high)
   fig = px.scatter(
       df[mask], x="sepal_width", y="sepal_length",
       color="species", size='petal_length',
       hover_data=['petal_width'])
   return fig

app.run_server()
```

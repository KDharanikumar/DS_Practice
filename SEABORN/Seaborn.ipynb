```python
import seaborn as sns
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
dharan = sns.load_dataset("iris")
print(dharan)

# Plot the responses for different events and regions
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=dharan)
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")

var=sns.load_dataset("diamonds")
print(var)
sns.lineplot(x="carat",y="price", hue="color",data=var)
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")

var=sns.load_dataset("penguins")
print(var)

sns.lineplot(x="flipper_length_mm",y="body_mass_g",hue="island", data=var)
# sns.scatterplot(y="flipper_length_mm",x="body_mass_g",hue="island",data=var)
```

```python
sns.displot(var,x="species",col="island",row="sex",binwidth=2,hue="species")
```

```python
sns.displot(var,x="flipper_length_mm",col="species",row="island",binwidth=1,hue="sex")
```

```python
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")

# Load the example tips dataset
tips = sns.load_dataset("tips")
print(tips)

# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="sales", y="total_bill",hue="smoker", palette=["m", "g"], data=tips)

# sns.boxplot(x="size", y="tip",
#             hue="sex", palette=["m", "g"],
#             data=tips)
# sns.despine(offset=20, left=False,right=False,trim=True)

# sns.boxplot(x="size", y="tip",hue="time", palette=["m", "g"], data=tips)
sns.despine(offset=20, left=False,right=False,trim=True)
```
